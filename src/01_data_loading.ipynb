The files train.csv and test.csv contain all the training samples as comma-sparated values. There are 3 columns in them, corresponding to class index (1 to 4), title and description. The title and description are escaped using double quotes ("), and any internal double quote is escaped by 2 double quotes (""). New lines are escaped by a backslash followed with an "n" character, that is "\n".

```python
import pandas as pd 

# load dataset 
train_path= "../data/raw/train.csv"

df=pd.read_csv(train_path)

df.head()
```

02- Data preproceess


```python
df["text"]= df["Title"] + " " + df["Description"]

df["labels"] = df["Class Index"]-1 #pytorch takes indx from 0

df= df[["text","labels"]]
df.head(10)
```

```python
df["text"] = df["text"].str.replace("\\n", " ", regex=True)
df["text"] = df["text"].str.strip()
```

```python

df.head(10)
```

Tokenising the text now

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
def tokenize_function(example):
    return tokenizer(
        example,
        padding="max_length",
        truncation=True,
        max_length=128
    )

# apply to all rows
tokenized = df["text"].apply(tokenize_function)
```

```python
from datasets import Dataset

# Convert pandas DataFrame to Dataset, which is compatible for training
dataset = Dataset.from_pandas(df)

def tokenize_function(exm):
    return tokenizer(
        exm["text"],   #  for Dataset
        padding="max_length",
        truncation=True,
        max_length=128
    )

dataset = dataset.map(tokenize_function, batched=True)
```

```python
dataset[0]
```

```python
dataset = dataset.remove_columns(["token_type_ids","text"])
dataset.set_format("torch")
```
